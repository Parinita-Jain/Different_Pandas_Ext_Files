```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("data2.csv")
df
```

```python
#step1 -: replace missing value with NaN
df["Python"].replace("missing",np.nan, inplace=True)

df["Machine Learning"].replace("?", np.nan, inplace=True)

df["Age"].replace("missing", np.nan, inplace=True)
df["Age"].replace("?", np.nan, inplace=True)

df["AI"].replace("missing", np.nan, inplace=True)

df["NLP"].replace("?", np.nan,inplace=True)

df["Cv"].replace("?", np.nan, inplace=True)
df["Cv"].replace("missing", np.nan, inplace=True)
```

```python
#step2-: Change the Datatype
df["Python"]=df["Python"].astype("float64")

df["Machine Learning"] = df["Machine Learning"].astype("float64")

df["Age"] = df["Age"].astype("float64")

df["AI"] = df["AI"].astype("float64")

df["NLP"] = df["NLP"].astype("float64")

df["Cv"]  =df["Cv"].astype("float64")
```

```python
#step3-: get the mean
pmean = df["Python"].mean()

mlmean =df["Machine Learning"].mean()

amean = df["Age"].mean()

aimean = df["AI"].mean()

nmean =df["NLP"].mean()

cmean = df["Cv"].mean()
```

```python
#step4 -: Use fillna
df["Python"].fillna(pmean, inplace=True)

df["Machine Learning"].fillna(mlmean, inplace=True)

df["Age"].fillna(amean, inplace=True)

df["AI"].fillna(aimean, inplace=True)

df["NLP"].fillna(nmean,inplace=True)

df["Cv"].fillna(cmean, inplace=True)
```
