```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```

Line Chart

```python
x=np.array([1,2,3,4,5])
y=x
plt.plot(x,y)
```

```python
x=np.array([1,2,3,4,5,6,7])
y=np.array([1,2,3,2,5,6,7])

plt.plot(x,y)
plt.xlabel("Number of year")
plt.ylabel("Growth of company")
plt.title("Growth w.r.t year")
plt.grid(True)
plt.show()
```

```python
x=np.array([600,800,1200,2200])
y=np.array([15,20,22,32])

plt.plot(x,y)
plt.xlabel("Area in sq-ft")
plt.ylabel("Price in Lakhs")
plt.title("Area w.r.t  Price")
plt.grid(True)
plt.show()
```

```python
x=np.array([600,800,1200,2200])
y=np.array([15,20,22,32])

plt.plot(x,y)
plt.xlabel("Area in sq-ft")
plt.ylabel("Price in Lakhs")
plt.title("Area w.r.t  Price")

plt.xlim(400,3000)
plt.ylim(10,50)

plt.grid(True)
plt.show()
```

```python
x=np.arange(1,10,0.1)
y=np.tan(x)

plt.plot(x,y)
plt.title("Sine wave")
plt.grid(True)
plt.show()
```

```python
x=np.arange(0,10,0.1)
y=np.sqrt(x)
y1=np.log(x)

plt.plot(x,y,color='red',label="sqrt")
plt.plot(x,y1,color='green',label="log")

plt.grid(True)
plt.legend()
plt.show()
```

Histogram plot

D/b histogram and bar plot-


1.  There are gaps between bars in a bar graph but in the histogram, the bars are adjacent to each other. 

2.  In histograms elements are grouped together, so that they are considered as ranges but in bar chart,elements are taken as individual entities.

3.  In histogram, Width of bars	Need not to be same, but in bar chart, width of bars are same.


4. Histogram show the frequency of numerical data,while, bar chart compare different categories of data.



```python
plt.hist(population)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Title")
plt.grid()
plt.show()
```

There are no bins here . 

```python
bins=[0,15,30,45,60,75,90]
plt.hist(population,bins)

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Title")
plt.grid()
plt.show()
```
