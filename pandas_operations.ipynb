```python
def getgrade(marks):
    v=marks/2
    if v>=75:
        return "A"
    elif v>=60 and v<75:
        return "B"
    elif v>=35 and v<60:
        return "C"
    else:
        return "F"

df["Grade"]=df["Total"].apply(getgrade)
df
```
